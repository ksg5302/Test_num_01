```python
import numpy as np
import requests as rq
import pandas as pd

import pymysql
from sqlalchemy import create_engine
pymysql.install_as_MySQLdb()
import MySQLdb


def make_csv(code_list):
    headers = {'User-agent':
               "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"} 
    df = pd.DataFrame()
    try:
        for i in range(1, 4):
            url = 'https://finance.naver.com/item/sise_day.nhn?code={0}&page='.format(code_list) + str(i)
            table = pd.read_html(rq.get(url,headers=headers).text,encoding="cp949")
            df_temp = table[0]
            df = df.append(df_temp)
        stock = df.dropna()
        return stock
    except:
        return stock

######################################################################################
listed_comp_df = pd.read_html('http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13', 
                                header=0)[0]

df = pd.DataFrame(listed_comp_df, columns=['종목코드'])

target = df['종목코드']
chart_list = []

for i in target:
    a = format(i, '06')
    chart_list.append(a)

df = pd.DataFrame()

for i in chart_list:
    df = make_csv(i)
    engine = create_engine("mysql+mysqldb://", encoding='utf-8')
    conn = engine.connect()
    df.to_sql(name='finance', con=engine, if_exists='append', index=False)
    conn.close()
```

# 전처리 순서

```python
from sklearn import preprocessing
import pandas as pd
```

```python
def user_fillna(df):
    df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
    df['Cabin'].fillna('N', inplace=True)
    df['Embarked'].fillna('N',inplace=True)
    df['Fare'].fillna(0, inplace = True)
    return df
```

```python
def user_drop(df):
    df.drop(['PassengerId','Name','Ticket'], axis=1,inplace=True)
    return df
```

```python
def user_format_features(df):
    features = ['Cabin', 'Sex', 'Embarked']
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df
```

```python
def transform_features(df):
    df = user_fillna(df)
    df = user_drop(df)
    df = user_format_features(df)
    return df
```

```python
titanic_df = pd.read_csv('train.csv')
y_titanic_df = titanic_df.pop('Survived')
x_titanic_df = titanic_df

x_titanic_df = transform_features(x_titanic_df)
```

## 스플릿 후 분석

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =train_test_split(x_titanic_df, y_titanic_df, test_size= 0.2, random_state = 11)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

dt_clf = DecisionTreeClassifier(random_state=11)
rf_clf = RandomForestClassifier(random_state=11)
lr_clf = LogisticRegression()

dt_clf.fit(x_train,y_train)
dt_pred = dt_clf.predict(x_test)
print('DecisionTreeClassifier 예측 정확도 : {0: .4f}'.format(accuracy_score(y_test,dt_pred)))

rf_clf.fit(x_train,y_train)
rf_pred = rf_clf.predict(x_test)
print('RandomForestClassifier 예측 정확도 : {0: .4f}'.format(accuracy_score(y_test,rf_pred)))

lr_clf.fit(x_train,y_train)
lr_pred = lr_clf.predict(x_test)
print('LogisticRegression 예측 정확도 : {0: .4f}'.format(accuracy_score(y_test,lr_pred)))
```

```python
from sklearn.model_selection import KFold
import numpy as np

features = x_titanic_df.to_numpy()
label = y_titanic_df.to_numpy()
dt_clf = DecisionTreeClassifier(random_state=156)
kfold = KFold(n_splits=5)
cv_accuracy = []

n_iter = 0

for train_index, test_index in kfold.split(features):
    
    x_train, x_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]

    dt_clf.fit(x_train,y_train)
    pred = dt_clf.predict(x_test)
    n_iter += 1

    accuracy = np.round(accuracy_score(y_test, pred), 4)
    train_size = x_train.shape[0]
    test_size = x_test.shape[0]

    print("\n#{0} 교차 검증 정확도 : {1}, 학습 데이터크기 : {2}, 검증데이터크기: {3}".format(n_iter,accuracy, train_size, test_size))
    # print("#{0} 검증 세트 인덱스 : {1}".format(n_iter, test_index))
    cv_accuracy.append(accuracy)

print("\n# 평균 검증 정확도 : ", np.mean(cv_accuracy))
```

```python
def exec_kfold(clf, folds = 5):
    kfold = KFold(n_splits=folds)
    cv_accuracy = []


    for n_iter, (train_index, test_index) in enumerate(kfold.split(x_titanic_df)):
        
        x_train, x_test = x_titanic_df.values[train_index], x_titanic_df.values[test_index]
        y_train, y_test = y_titanic_df.values[train_index], y_titanic_df.values[test_index]

        dt_clf.fit(x_train,y_train)
        pred = dt_clf.predict(x_test)

        accuracy = np.round(accuracy_score(y_test, pred), 4)
        train_size = x_train.shape[0]
        test_size = x_test.shape[0]

        print("\n#{0} 교차 검증 정확도 : {1}, 학습 데이터크기 : {2}, 검증데이터크기: {3}".format(n_iter+1,accuracy, train_size, test_size))
        # print("#{0} 검증 세트 인덱스 : {1}".format(n_iter, test_index))
        cv_accuracy.append(accuracy)

    print("\n# 평균 검증 정확도 : ", np.mean(cv_accuracy))


from sklearn.model_selection import KFold
import numpy as np

exec_kfold(dt_clf, folds = 5)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'max_depth':[2,3,5,10,15,20], 'min_samples_split' : [2,3,5], 'min_samples_leaf':[1,5,8]}

grid_dtclf = GridSearchCV(dt_clf, param_grid = parameters, scoring='accuracy', cv=5)
grid_dtclf.fit(x_train,y_train)

print('최적 dt_clf 하이퍼 파라미터: ', grid_dtclf.best_params_)
print('최고 dt_clf 정확도 : {:.4f}'.format(grid_dtclf.best_score_))
best_dtclf = grid_dtclf.best_estimator_

dpred = best_dtclf.predict(x_test)
print('dt_clf 정확도 : {:.4f}'.format(accuracy_score(y_test, dpred)))
print()

grid_rfclf = GridSearchCV(rf_clf, param_grid = parameters, scoring='accuracy', cv=5)
grid_rfclf.fit(x_train,y_train)

print('최적 rf_clf 하이퍼 파라미터: ', grid_rfclf.best_params_)
print('최고 rf_clf 정확도 : {:.4f}'.format(grid_rfclf.best_score_))
best_rfclf = grid_rfclf.best_estimator_

rfpred = best_rfclf.predict(x_test)
print('rf_clf 정확도 : {:.4f}'.format(accuracy_score(y_test, rfpred)))
print()

# grid_lrclf = GridSearchCV(lr_clf, param_grid = parameters, scoring='accuracy', cv=5)
# grid_lrclf.fit(x_train, y_train)

# print('최적 lr_clf 하이퍼 파라미터: ', grid_lrclf.best_params_)
# print('최고 lr_clf 정확도 : {:.4f}'.format(grid_lrclf.best_score_))
# best_lrclf = grid_lrclf.best_estimator_

# lrpred = best_lrclf.predict(x_test)
# print('lr_clf 정확도 : {:.4f}'.format(accuracy_score(y_test, lrpred)))
# print()
```
